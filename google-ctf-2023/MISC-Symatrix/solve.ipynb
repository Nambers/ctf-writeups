```python
from PIL import Image 
```

```python
base_image = Image.open("./symatrix.png")
x_len, y_len = base_image.size
base_matrix = base_image.load()
```

```python
pixelDict = {}
nx_len = x_len - 1
for x in range(0, x_len):
    for y in range(0, y_len):
        r, g, b = base_matrix[nx_len - x, y][0], base_matrix[nx_len - x, y][1], base_matrix[nx_len - x, y][2]
        if (r, g, b) not in pixelDict:
            pixelDict[(r, g, b)] = 1
        else:
            pixelDict[(r, g, b)] += 1
```

```python
print(sorted(pixelDict.items(), key=lambda x: x[1]))
```

```python
print(pixelDict[(0, 1, 0)])
```

```python
nx_len = x_len - 1
bstr = ""
lastPos = 0
for y in range(0, y_len):
    for x in range(0, x_len):
        r, g, b = base_matrix[nx_len - x, y][0], base_matrix[nx_len - x, y][1], base_matrix[nx_len - x, y][2]
        if b == 1:
            bstr += "1"
        elif g  > 0 and (pixelDict[(r, g, b)] <= 300 or (r, g, b) == (0, 1, 0)):
            print((r, g, b))
            bstr += "0"
print(bstr)
```

```python
def decode_binary_string(s):
    return ''.join(chr(int(s[i*8:i*8+8],2)) for i in range(len(s)//8))
print(decode_binary_string(bstr))
```
